```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
# Load the dataset
file_path = '4colorsresult.csv'  # Replace with your actual file path
df = pd.read_csv(file_path)

# Remove empty rows
df = df.dropna(how='all')

# Display the first few rows of the DataFrame
df.head()
```

```python
# Scatter plot for Actual Luminance (g1) vs Final Angular Velocity
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.regplot(x='Actual Luminance (g1)', y='Final Angular Velocity', data=df, ci=None, scatter_kws={'s': 50})
plt.title('Actual Luminance (g1) vs Final Angular Velocity')
plt.xlabel('Actual Luminance (g1)')
plt.ylabel('Final Angular Velocity')

plt.show()
```

```python
# Group by RGB (g1) and calculate the mean Final Angular Velocity
grouped_data = df.groupby('RGB (g1)')['Final Angular Velocity'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='RGB (g1)', y='Final Angular Velocity', data=grouped_data)
plt.title('Average Final Angular Velocity by RGB (g1)')
plt.xlabel('RGB (g1)')
plt.ylabel('Average Final Angular Velocity')
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Selecting numerical features for PCA
features = ['L value (g1)', 'Actual Luminance (g1)', 'L value (g2)', 'Actual Luminance (g2)', 'Final Angular Velocity']
x = df[features]

# Standardizing the features
x = StandardScaler().fit_transform(x)

# Performing PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x)

# Creating a DataFrame with the principal components
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Plotting the PCA result
plt.figure(figsize=(10, 8))
plt.scatter(principal_df['PC1'], principal_df['PC2'], c=df['Final Angular Velocity'], cmap='viridis')
plt.colorbar(label='Final Angular Velocity')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of the Dataset')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Extracting the 'a', 'b', 'L', and 'Final Angular Velocity' components from Lab (g1)
df['a (g1)'] = df['Lab (g1)'].apply(lambda x: float(x.strip('()').split(',')[1]))
df['b (g1)'] = df['Lab (g1)'].apply(lambda x: float(x.strip('()').split(',')[2]))
df['L (g1)'] = df['L value (g1)']

# Plotting
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df['a (g1)'], df['b (g1)'], 
                      s=df['L (g1)']*10, # Multiplying by 10 to make the size more visible
                      c=df['Final Angular Velocity'], cmap='viridis', alpha=0.6, edgecolors="w", linewidth=0.5)

# Adding a color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Final Angular Velocity')

# Setting labels and title
plt.xlabel('a (g1)')
plt.ylabel('b (g1)')
plt.title('Lab Color Representation with Final Angular Velocity')

# Adding grid for better visualization
plt.grid(True)

# Displaying the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np



# Split the Lab (g1) and Lab (g2) values into separate columns for L, a, b
df[['L1', 'a1', 'b1']] = df['Lab (g1)'].str.extract(r'\(([^,]+), ([^,]+), ([^,]+)\)').astype(float)
df[['L2', 'a2', 'b2']] = df['Lab (g2)'].str.extract(r'\(([^,]+), ([^,]+), ([^,]+)\)').astype(float)

# Calculate color contrast using Euclidean distance in Lab space
df['Color Contrast'] = np.sqrt((df['L1'] - df['L2'])**2 + (df['a1'] - df['a2'])**2 + (df['b1'] - df['b2'])**2)

# Create the scatter plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['Color Contrast'], df['Final Angular Velocity'],
s=df['Color Contrast']*3,
c=df['Final Angular Velocity'], cmap='coolwarm', alpha=0.6, edgecolors="w", linewidth=0.5)

# Adding a color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Final Angular Velocity')

# Adding labels and title
plt.xlabel('Color Contrast')
plt.ylabel('Final Angular Velocity')
plt.title('Color Contrast vs. Final Angular Velocity')

# Adding grid for better visualization
plt.grid(True)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load your CSV data into a DataFrame
df = pd.read_csv('4colorsresult.csv')

# Define bins for actual luminance values
bins = np.linspace(df['Actual Luminance (g1)'].min(), df['Actual Luminance (g1)'].max(), 20)
df['Luminance Bin'] = pd.cut(df['Actual Luminance (g1)'], bins)

# Group the data by the luminance bin
grouped = df.groupby('Luminance Bin', observed=False)['Final Angular Velocity']

# Prepare data for box plot
data_for_box_plot = [group for _, group in grouped]

# Get the positions for the box plots (use the midpoint of each bin for the x-axis)
positions = [bin.mid for bin in grouped.groups.keys()]

# Calculate mean and standard deviation of the angular velocities
mean_velocity = df['Final Angular Velocity'].mean()
std_velocity = df['Final Angular Velocity'].std()

# Define the threshold for non-significant data (within one standard deviation)
threshold_upper = mean_velocity + std_velocity
threshold_lower = mean_velocity - std_velocity

# Plot the box plot
plt.figure(figsize=(12, 6))
plt.boxplot(data_for_box_plot, positions=positions, widths=2, patch_artist=True)

# Add horizontal lines for the threshold
plt.axhline(y=threshold_upper, color='r', linestyle='--', label=f'Upper Threshold ({threshold_upper:.2f})')
plt.axhline(y=threshold_lower, color='r', linestyle='--', label=f'Lower Threshold ({threshold_lower:.2f})')

# Customize the plot
plt.xlabel('Actual Luminance Value')
plt.ylabel('Angular Velocity')
plt.title('Box Plot of Angular Velocity vs Actual Luminance Value with Significant Levels')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML

# Load your CSV data into a DataFrame
df = pd.read_csv('4colorsresult.csv')

# Define bins for actual luminance values
bins = np.linspace(df['Actual Luminance (g1)'].min(), df['Actual Luminance (g1)'].max(), 20)
df['Luminance Bin'] = pd.cut(df['Actual Luminance (g1)'], bins)

# Calculate mean and standard deviation of the angular velocities
mean_velocity = df['Final Angular Velocity'].mean()
std_velocity = df['Final Angular Velocity'].std()

# Define the threshold for non-significant data (within one standard deviation)
threshold_upper = mean_velocity + std_velocity
threshold_lower = mean_velocity - std_velocity

# Filter the data to find entries outside the thresholds
outside_threshold = df[(df['Final Angular Velocity'] > threshold_upper) | (df['Final Angular Velocity'] < threshold_lower)]

# Extract relevant color information for g1 and g2
outside_threshold_colors = outside_threshold[['Lab (g1)', 'Lab (g2)', 'Final Angular Velocity']].copy()

# Function to calculate Euclidean distance in Lab space (CIE76)
def calculate_cie76_distance(lab1, lab2):
    return np.sqrt(np.sum((np.array(lab1) - np.array(lab2)) ** 2))

# Calculate contrast for g1 with black and g2 with white
black_lab = [0, 0, 0]
white_lab = [100, 0, 0]
outside_threshold_colors['Contrast (g1, Black)'] = outside_threshold_colors['Lab (g1)'].apply(lambda x: calculate_cie76_distance(eval(x), black_lab))
outside_threshold_colors['Contrast (g2, White)'] = outside_threshold_colors['Lab (g2)'].apply(lambda x: calculate_cie76_distance(eval(x), white_lab))

# Calculate the difference between g1 and g2
outside_threshold_colors['Difference (g1, g2)'] = outside_threshold_colors.apply(
    lambda row: calculate_cie76_distance(eval(row['Lab (g1)']), eval(row['Lab (g2)'])), axis=1
)

# Prepare final DataFrame for display
outside_threshold_colors_display = outside_threshold_colors[['Lab (g1)', 'Contrast (g1, Black)', 'Lab (g2)', 'Contrast (g2, White)', 'Difference (g1, g2)', 'Final Angular Velocity']]

# Generate HTML table with custom styling
html_table = outside_threshold_colors_display.to_html(index=False, table_id="data_table", classes="table table-striped table-bordered")

# Additional CSS styling
html_style = """
<style>
#data_table thead th {
    background-color: #f2f2f2;
    font-weight: bold;
}
#data_table tbody tr:nth-child(even) {
    background-color: #f9f9f9;
}
#data_table tbody tr:nth-child(odd) {
    background-color: white;
}
</style>
"""

# Display the styled table with caption
html_output = f"""
{html_style}
<h2>Data Points Outside Threshold with Color Contrast and Difference</h2>
{html_table}
"""
display(HTML(html_output))
```
